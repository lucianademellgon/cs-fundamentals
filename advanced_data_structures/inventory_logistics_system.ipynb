## ğŸ“‹ Project Overview
A simulation of warehouse logistics using **Abstract Data Types (ADTs)** to ensure data integrity and uniqueness.
This module addresses the **"Duplicate Asset Problem"** in supply chains.

### âš™ï¸ Technical Specs
* **Core Logic:** Custom implementation of `__hash__` and `__eq__` methods.
* **Data Structure:** Python `Set` (Hash Map implementation).
* **Complexity:**
    * Insertion: O(1) average case.
    * Lookup: O(1) average case.

### ğŸ” Key Implementation
The `HardwareUnit` class overrides default object comparison to ensure that two physical assets with the same specifications are treated as the same entity in the system, preventing double-counting in the `WarehouseInventory`.

```python
class HardwareUnit:
  #Represents a physical asset in the warehouse.
  #Implements Hashable protocol to allow storage in Sets.
  def __init__(self, brand, model, defect_code):
    self.brand = brand
    self.model = model
    self.defect_code = defect_code  # 0: Perfect, 1-5: Severity of defect

  # --- METHODS FOR HASHING ---
  # These are crucial for the Set data structure to work correctly.

  def __eq__(self, other):
    # Determines if two objects are 'identical' in business terms.
    if not isinstance(other, HardwareUnit):
      return False
    return (self.brand == other.brand and self.model == other.model and self.defect_code == other.defect_code)

  def __hash__(self):
    # Generates a unique hash based on content, not memory address. Allows O(1) lookup in Sets.
    return hash((self.brand, self.model, self.defect_code))

  def __str__(self):
    return f"ğŸ’» Asset: {self.brand} {self.model} | Condition Code: {self.defect_code}"


class WarehouseInventory:
    # Manager class using a Set ADT to enforce uniqueness.
  def __init__(self):
    self.storage = set()  # The Set guarantees no duplicates

  def add_asset(self, asset):
    # Attempts to add an item to inventory. Rejects duplicates.
    if asset in self.storage:
      # This check is O(1) thanks to the hash map
      print(f"âš ï¸  ALERT: Asset '{asset.brand} {asset.model}' is already registered.")
    else:
      self.storage.add(asset)
      print(f"âœ… SUCCESS: {asset.brand} {asset.model} added to inventory.")

  def check_availability(self, asset):
    return asset in self.storage

  def generate_report(self):
    if not self.storage:
      print("ğŸ“¦ [SYSTEM]: Warehouse is empty.")
    else:
      print(f"\nğŸ“‹ --- WAREHOUSE STATUS REPORT ({len(self.storage)} items) ---")
      for item in self.storage:
        print(f"   - {item}")
        print("--------------------------------------------------")


# --- Unit Testing / Simulation ---
if __name__ == "__main__":
  depot = WarehouseInventory()

  print("--- 1. Ingesting New Assets ---")
  # Creating asset instances
  unit_1 = HardwareUnit("Dell", "Inspiron 15", 2)
  unit_2 = HardwareUnit("HP", "Pavilion", 1)
  unit_3 = HardwareUnit("Lenovo", "ThinkPad", 3)
  unit_4_duplicate = HardwareUnit("Dell", "Inspiron 15", 2)  # Identical to unit_1

  # Adding to system
  depot.add_asset(unit_1)
  depot.add_asset(unit_2)
  depot.add_asset(unit_3)

  print("\n--- 2. Testing Duplicate Prevention ---")
  depot.add_asset(unit_4_duplicate)  # Should trigger alert

  # Final Report
  depot.generate_report()

  # Verification
  print("\n--- 3. Audit Check ---")
  query = HardwareUnit("HP", "Pavilion", 1)
  is_present = depot.check_availability(query)
  print(f"Is 'HP Pavilion' in stock? -> {is_present}")
```
