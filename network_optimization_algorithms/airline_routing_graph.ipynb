## ğŸ“‹ Project Overview
A routing system designed to model global flight connectivity. This module uses a **Directed Graph** architecture where airports are nodes and flight paths are weighted edges.
It demonstrates fundamental algorithms for pathfinding and network traversal.

### âš™ï¸ Technical Specs
* **Data Structure:** Adjacency List (Implemented via Python Dictionaries).
* **Graph Type:** Directed & Weighted (Flights have direction and duration/cost).
* **Key Algorithms:**
    * **DFS (Depth-First Search):** For connectivity checks.
    * **Pathfinding:** Logic to determine route availability.

### ğŸ” Usage
The `FlightNetwork` class allows for dynamic route establishment and queries like *"Is there a connection between JFK and Tokyo?"*, simulating the backend logic of travel booking engines.

```python
class AirportNode:
  # Represents a Node in the Graph (An Airport).
  def __init__(self, code, city):
    self.code = code  # IATA Code (e.g., JFK, LHR)
    self.city = city

  def __str__(self):
    return f"{self.code} ({self.city})"

  def __repr__(self):
    return self.code

class FlightNetwork:
  # Graph implementation using an Adjacency List.
  # Manages connections (Edges) between Airports (Nodes).
  def __init__(self):
    # Dictionary to store the graph: { "JFK": [("LHR", 7h), ("LAX", 6h)] }
    self.routes = {}
    self.airports = {} # Registry of created airport objects

  def add_airport(self, code, city):
    # Adds a Node to the graph.
    if code not in self.airports:
      new_airport = AirportNode(code, city)
      self.airports[code] = new_airport
      self.routes[code] = [] # Initialize empty connections list
      print(f"ğŸ“ Airport Added: {new_airport}")

  def add_flight(self, origin_code, dest_code, duration_hrs):
    # Adds a Directed Edge with a Weight (Duration). Origin -> Destination
    if origin_code in self.routes and dest_code in self.routes:
      # Append tuple (Destination, Weight)
      self.routes[origin_code].append( (dest_code, duration_hrs) )
      print(f"âœˆï¸  Route Created: {origin_code} â {dest_code} ({duration_hrs}h)")
    else:
      print(f"âš ï¸ Error: One of the airports {origin_code}/{dest_code} does not exist.")

  def show_connections(self):
    # Visualizes the Adjacency List.
    print("\n--- ğŸŒ GLOBAL FLIGHT NETWORK ---")
    for origin, destinations in self.routes.items():
      if not destinations:
        print(f"ğŸ›« {origin}: No outbound flights.")
      else:
        # Formatting output for readability
          connections = [f"â {dest} ({time}h)" for dest, time in destinations]
          print(f"ğŸ›« {origin}: {', '.join(connections)}")
    print("--------------------------------\n")

  def find_route(self, start_code, end_code, path=None):
    # DFS (Depth First Search) Algorithm to check connectivity. Returns the path if a route exists, None otherwise.
    if path is None:
      path = []
      path = path + [start_code]
      if start_code == end_code:
        return path
        if start_code not in self.routes:
          return None
        # Recursive search through neighbors
        for connection in self.routes[start_code]:
          destination = connection[0] # Get the code from tuple (code, time)
          if destination not in path:
            new_path = self.find_route(destination, end_code, path)
            if new_path:
              return new_path
      return None

# --- System Simulation ---
if __name__ == "__main__":
  network = FlightNetwork()

  # 1. Registering Airports (Nodes)
  print("--- 1. Initializing Network ---")
  network.add_airport("JFK", "New York")
  network.add_airport("LHR", "London")
  network.add_airport("CDG", "Paris")
  network.add_airport("HND", "Tokyo")
  network.add_airport("SYD", "Sydney")
  print("--------------------------------")

  # 2. Creating Routes (Edges)
  # Note: These are directed. JFK->LHR doesn't automatically mean LHR->JFK
  network.add_flight("JFK", "LHR", 7)
  network.add_flight("LHR", "CDG", 1)
  network.add_flight("LHR", "HND", 12)
  network.add_flight("HND", "SYD", 9)
  network.add_flight("JFK", "CDG", 8) # Direct flight

  # 3. Display Graph
  network.show_connections()

  # 4. Testing Pathfinding Algorithm (DFS)
  print("--- 2. Route Search Engine ---")

  # Case A: Route exists (New York -> Sydney via London & Tokyo)
  search_1 = ("JFK", "SYD")
  route_1 = network.find_route(*search_1)
  print(f"ğŸ” Searching path from {search_1[0]} to {search_1[1]}...")
  if route_1:
    print(f"âœ… Route Found: {' â '.join(route_1)}")
  else:
    print("âŒ No connection available.")

  print()

  # Case B: Route does not exist (Sydney -> New York, assuming no return flights added)
  search_2 = ("SYD", "JFK")
  route_2 = network.find_route(*search_2)
  print(f"ğŸ” Searching path from {search_2[0]} to {search_2[1]}...")
  if route_2:
    print(f"âœ… Route Found: {' â '.join(route_2)}")
  else:
    print("âŒ No connection available (One-way flights logic).")
```
